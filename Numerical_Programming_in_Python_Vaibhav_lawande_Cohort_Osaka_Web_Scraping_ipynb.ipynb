```python
#import Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
# 1.Source Selection:

responce = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=it&txtLocation=India")
soup=BeautifulSoup(responce.text,"html.parser")
soup
```

```python
#2.Data Extraction:
job_data=soup.find("ul",class_="new-joblist")

job_data
```

```python
job_data1=job_data.find_all("li",class_="clearfix job-bx wht-shd-bx")
job_data1
```

```python
#extract Company name
l=[]
company=[]
for i in job_data1:
  k=i.find('h3',class_="joblist-comp-name")
  l.append(k.text.strip())
for i in l :
  company.append(i.replace('\r\n     (More Jobs)',""))

company
```

```python
#extract Job Role
role=[]
for i in job_data1:
  k=i.find('a')
  role.append(k.text.strip())

role
```

```python
#extract experience

experience=[]
for i in job_data1:
  k=i.find('i',class_="material-icons").next_sibling
  experience.append(k.text)

experience
```

```python
#extract package

package=[]
for i in job_data1:
  try:
    k=i.find('i',class_="material-icons rupee").next_sibling
  except AttributeError:
    k = None
  package.append(k)

package
```

```python
#extract Location

location=[]
for i in job_data1:
  k=i.find_all('span')[1].text
  location.append(k)

location
```

```python
#create dictionary to merger all the out put in single frame

dict1={"Company Name":company,"Role":role,"Experience":experience,"Salary":package,"Job Location":location}
dict1
```

```python
#create Data frame
job_df=pd.DataFrame(dict1)
job_df
```

```python
# To save in csv file!
job_df.to_csv("job_details.csv")
```
